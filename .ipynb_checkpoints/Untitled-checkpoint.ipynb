```python
#import required packages
import pandas as pd
import numpy as np
#read data 
dataset=pd.read_csv('TransactionClassificationDataSet.csv', sep=';',index_col=0)
```

```python
# Import the count vectorizer and initialize it
from sklearn.feature_extraction.text import CountVectorizer

# We use CountVectorizer to split up each email message into a list of words and their counts.
count_vector = CountVectorizer()

#https://imguploader.com/image/bild-2021-05-05-170628.iz8p
```

```python
combined= dataset['Buchungstext']+' '+dataset['Beguenstigter/Zahlungspflichtiger']+' '+dataset['Verwendungszweck']
counts = count_vector.fit_transform(combined).toarray()
count_vector.fit(combined)
names = count_vector.get_feature_names()
print(counts)
```

```python
doc_array = count_vector.transform(combined).toarray()
doc_array=pd.DataFrame(data=doc_array)
doc_array
```

```python
frequency_matrix = pd.DataFrame(data=doc_array, columns=names,index= dataset.index)
#frequency_matrix
c=pd.DataFrame(dataset['Betrag'])
data1 = c.join(frequency_matrix)
data1
```
